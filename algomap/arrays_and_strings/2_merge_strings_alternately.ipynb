# Leetcode 1768
https://leetcode.com/problems/merge-strings-alternately/description/

**You are given two strings word1 and word2. Merge the strings by adding letters in alternating order, starting with word1. If a string is longer than the other, append the additional letters onto the end of the merged string. Return the merged string.**

### Example 1

**Input**: word1 = "abc", word2 = "pqr"

**Output**: "apbqcr"

**Explanation**: The merged string will be merged as so:

word1:  a   b   c

word2:    p   q   r

merged: a p b q c r

```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        merged = '' # O(1) initially, but grows to O(n+m)
        for i in range(min(len(word1), len(word2))): # O(min(n,m))
            merged += word1[i] + word2[i] # O(1) per operation, but O(n+m) total due to string immutability

        merged += word1[i + 1:] # O(n)
        merged += word2[i + 1:] # O(m)

        return merged
    
# Time: O(n+m) where n = len(word1) and m = len(word2)
# Space: O(n+m) for the merged string
    
s = Solution()
print(s.mergeAlternately('abc', 'pqr'))
```
