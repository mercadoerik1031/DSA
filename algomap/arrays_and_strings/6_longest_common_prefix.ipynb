# Leetcode 14
https://leetcode.com/problems/longest-common-prefix/

**Write a function to find the longest common prefix string amongst an array of strings. If there is no common prefix, return an empty string "".**

### Example 1:

**Input**: strs = ["flower","flow","flight"]

**Output**: "fl"

### Example 2:

**Input**: strs = ["dog","racecar","car"]

**Output**: ""

**Explanation**: There is no common prefix among the input strings.


```python
from typing import List

class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        prefix = strs[0] # Time: O(1), Space: O(1)

        for i in range(1, len(strs)): # Time: O(n), Space: 0(1)
            while not strs[i].startswith(prefix): # Time: O(m), Space: O(1) 
                prefix = prefix[:-1] # Time: O(1), Space: O(1)
            
            if not prefix: # Time: O(1), Space: O(1)
                return ''
            
        return prefix
    
# Time: O(n * m), Space: O(1)
# The outer for loop runs n-1 times where n is the number of string in the input list
# The inner while loop can potentially iterate up to m times in the worst case where m is the length
    # of the first string. This is because in the worst case prefix might need to remove all its chars
    
s = Solution()
print(s.longestCommonPrefix(["flower","flow","flight"]))
print(s.longestCommonPrefix(["dog","racecar","car"]))
```
