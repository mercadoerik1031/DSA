# Leetcode 15

https://leetcode.com/problems/3sum/

**Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.**

Notice that the solution set must not contain duplicate triplets.

### Example 1:

**Input**: nums = [-1,0,1,2,-1,-4]

**Output**: [[-1,-1,2],[-1,0,1]]

**Explanation**: 
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.

The distinct triplets are [-1,0,1] and [-1,-1,2].

Notice that the order of the output and the order of the triplets does not matter.

### Example 2:

**Input**: nums = [0,1,1]

**Output**: []

**Explanation**: The only possible triplet does not sum up to 0.

### Example 3:

**Input**: nums = [0,0,0]

**Output**: [[0,0,0]]

**Explanation**: The only possible triplet sums up to 0.

```python
from typing import List

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        # Sort nums to use two pointer technique
        nums.sort() # Time: O(n log(n)), Space: O(1)
        result = [] # Time: O(1), SpaceL O(1)

        for i in range(len(nums)): # Time: O(n), Space: O(1)
            # Check if i > 0 and if pervious num is equal to current num
            if i > 0 and nums[i] == nums[i - 1]: # Time: O(1), Space: O(1)
                continue

            # initialize 2 pointers
            j = i + 1 # Time: O(1), Space: O(1)
            k = len(nums) - 1 # Time: O(1), Space: O(1)

            # Loop while second pointer is less then third pointer
            while j < k: # Time: O(n), Space: O(1)
                # Calc current sum
                cur_sum = nums[i] + nums[j] + nums[k] # Time: O(1), Space: O(1)

                # Check if sum less than 0
                if cur_sum < 0: # Time: O(1), Space: O(1)
                    # Move j pointer up
                    j += 1 # Time: O(1), Space: O(1)
                # Elif sum greater than 0
                elif cur_sum > 0: # Time: O(1), Space: O(1)
                    # Move k pointer down
                    k -= 1 # Time: O(1), Space: O(1)
                # Else sum is equal to 0 append result and increment j
                else:
                    result.append([nums[i], nums[j], nums[k]]) # Time: O(1), Space: O(1)
                    j += 1 # Time: O(1), Space: O(1)

                    # Increment j while there are duplicates
                    while nums[j - 1] == nums[j] and j < k: # Time: O(n), Space: O(1)
                        j += 1 # Time: O(1), Space: O(1)

        return result # Time: O(1), Space: O(1)
    
# Time: O(n^2), Space: O(1)
    
s = Solution()
print(s.threeSum([-1,0,1,2,-1,-4]))
print(s.threeSum([0,1,1]))
print(s.threeSum([0,0,0]))
```
