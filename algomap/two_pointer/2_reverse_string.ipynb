# Leetcode 344

https://leetcode.com/problems/reverse-string/description/

**Write a function that reverses a string. The input string is given as an array of characters s.**

**You must do this by modifying the input array in-place with O(1) extra memory.**

 

### Example 1:

**Input**: s = ["h","e","l","l","o"]

**Output**: ["o","l","l","e","h"]

### Example 2:

**Input**: s = ["H","a","n","n","a","h"]

**Output**: ["h","a","n","n","a","H"]
 

**Constraints**:

1 <= s.length <= 105
s[i] is a printable ascii character.

```python
from typing import List

class Solution:
    def reverseString(self, s: List[str]) -> None:
        left, right = 0, len(s) - 1 # Time: O(1), Space: O(1)

        while left <= right: # Time: O(n), Space: O(1). The while loop is n/2 because left and right move towards each other.
            s[left], s[right] = s[right], s[left] # Time: O(1), Space: O(1)
            left += 1 # Time: O(1), Space: O(1)
            right -= 1 # Time: O(1), Space: O(1)

# Time: O(n), Space: O(1)

s = Solution()

string_1 = ["h","e","l","l","o"]
s.reverseString(string_1)
print(string_1)

string_2 = ["H","a","n","n","a","h"]
s.reverseString(string_2)
print(string_2)
```
