```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        dic = {} # Time: O(1), Space: O(1) but dict can grow to O(n-1)

        for i in range(len(nums)): # Time: O(n), Space: O(1)
            composite = target - nums[i] # Time: O(1), Space: O(1)

            if nums[i] in dic: # Time: O(1), Space: O(1)
                return [dic[nums[i]], i] # Time: O(1), Space: O(1)
            else:
                dic[composite] = i # Time: O(1), Space: O(1)

        return [] # Time: O(1), Space: O(1)
    
# Time: O(n), Space: O(n)
# In worst case dic can grow to O(n - 1) therefore Space complexity is O(n)
    
s = Solution()
print(s.twoSum([2,7,11,15], 9))
print(s.twoSum([3,2,4], 6))
print(s.twoSum([3,3], 6))
print(s.twoSum([-1,-2,-3,-4,-5], -8))
```
