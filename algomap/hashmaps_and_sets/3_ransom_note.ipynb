# Leetcode 383

https://leetcode.com/problems/ransom-note/description/

**Given two strings ransomNote and magazine, return true if ransomNote can be constructed by using the letters from magazine and false otherwise.**

**Each letter in magazine can only be used once in ransomNote.**

 

### Example 1:

**Input**: ransomNote = "a", magazine = "b"

**Output**: false

### Example 2:

**Input**: ransomNote = "aa", magazine = "ab"

**Output**: false

### Example 3:

**Input**: ransomNote = "aa", magazine = "aab"

**Output**: true
 
**Constraints**:

1 <= ransomNote.length, magazine.length <= 105
ransomNote and magazine consist of lowercase English letters.

```python
from collections import Counter

class Solution:
    def canConstruct(self, randomNote: str, magazine: str) -> bool:
        r_counter = Counter(randomNote) # Time: O(n), Space: O(n)
        m_counter = Counter(magazine) # Time: O(m), Space: O(m)
        r_set = set(randomNote) # Time: O(n), Space: O(n)

        for char in r_set: # Time: O(n), Space: O(1)
            if char not in m_counter or m_counter[char] < r_counter[char]: # Time: O(1), O(1)
                return False
    
        return True
    
# Time: O(n), Space: O(n + m)
# Time complexity can be further simplified to O(n) as the len of ransomNote
# is the dominant factor. This is because the len of magazine 
# is only used in the constant-time dictionary operations, which do not
# contribute to the overall time complexity

s = Solution()
print(s.canConstruct('a', 'b'))
print(s.canConstruct('aa', 'ab'))
print(s.canConstruct('aa', 'aab'))
```
