# Problem 2
#### https://leetcode.com/problems/remove-linked-list-elements/submissions/1056505813/

Given the <code>head</code> of a linked list and an integer <code>val</code>, remove all the nodes of the linked list that has <code>Node.val == val</code>, and return the new <code>head</code>.

### 1. State the problem clearly. Identify the input & output formats.

Given the <code>head</code> of a linked list and a int <code>val</code> remove any node that contains <code>val</code>

Example:
Input

<code>head = [1, 2, 6, 3, 4, 5, 6]</code>

<code>val = 6</code>

Output:

<code>[1,2, 3, 4, 5]</code>

### 2. Come up with some example inputs & outputs. Try to cover all edge cases.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
test_cases = [
    (ListNode(1, ListNode(2, ListNode(6, ListNode(3, ListNode(4, ListNode(5, ListNode(6))))))), 6),
    (ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5))))), 6),  # No occurrences of 6
    (None, 6),  # Empty list
    (ListNode(6), 6),  # Single-node list with value 6
    (ListNode(1, ListNode(2, ListNode(3))), 6),  # No occurrences of 6
]
```

### 3. Come up with a solution

we create a holder node to help us look at the next node
we set current to holder and we begin looking at the next node

* if <code>current.next</code> is <code>None</code>
    * stop of the loop
* if the value of the next node = val
    * the next node = current nex next
        * example: <code>target = 7</code>; <code>list = 1,7,2</code>
            * <code>current.val = 1</code>; <code>current.next.val = 7</code>
                * because <code>current.next.val = target</code>
                * <code>current.next = current.next.next = 2</code>
* else
    * current = the next node

* return the node next of holder

```python
class Solution:
    def remove_elements(self, head, val):
        holder = ListNode()
        holder.next = head
        current = holder

        while current.next is not None:
            if current.next.val == val:
                current.next = current.next.next
            else:
                current = current.next

        return holder.next
```

#### Test Solution

```python
def run_tests():
    s = Solution()
    for idx, (input_list, val) in enumerate(test_cases, start=1):
        result = s.remove_elements(input_list, val)
        print(f"Test case {idx}:")
        print("Output:", list_values(result))
        print()
```

```python
# Helper function to convert ListNode to list for easier comparison
def list_values(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values
```

```python
run_tests()
```
