# Problem 1
#### https://leetcode.com/problems/convert-binary-number-in-a-linked-list-to-integer/description/

Given <code>head</code> which is a reference node to a singly-linked list. The value of each node in the linked list is either 0 or 1. The linked list holds the binary representation of a number.

Return the decimal value of the number in the linked list.

The most significant bit is at the head of the linked list.

### 1. State the problem clearly. Identify the input & output formats.

Given the start of a linked list traverse the list to find the binary value. Then convert it into decimal form.

Input:
<code>head</code> = the start of a linked list

Output:

<code>decimal</code> value of a binary number

### 2. Come up with some example inputs & outputs. Try to cover all edge cases.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
tests = [
    {"input": ListNode(1, ListNode(0, ListNode(1))), "output": 5},
    {"input": ListNode(1, ListNode(1, ListNode(0))), "output": 6},
    {"input": ListNode(0, ListNode(0, ListNode(0))), "output": 0},
    {"input": ListNode(1, ListNode(1, ListNode(1))), "output": 7},
    {"input": ListNode(0, ListNode(0, ListNode(1))), "output": 1},
    {"input": ListNode(1), "output": 1},  # Single-node list with value 1
    {"input": ListNode(0), "output": 0},  # Single-node list with value 0
    {"input": ListNode(1, ListNode(1, ListNode(1, ListNode(0)))), "output": 14},
    {"input": ListNode(1, ListNode(1, ListNode(0, ListNode(1, ListNode(1))))), "output": 27}
]
```

### 3. Come up with a solution

* we begin by making the head of the linked list equal to the <code>current</code> node
* we create an empty string variable <code>binary</code> to store the values of each node into a string
* while the <code>current</code> node we are working on is not equal to <code>None</code>
    * turn the value of <code>current</code> to a string and concatenate it to <code>binary</code>
    * <code>current</code> now equals the next node
* once <code>current</code> equals <code>None</code>
    * return the integer value of <code>binary</code> with base 2

```python
class Solution:
    def get_decimal_value(self, head: ListNode) -> int:
        current = head
        binary = ""

        while current is not None:
            binary += str(current.val)
            current = current.next

        return int(binary, 2)
```

#### Test Solution

```python
solution = Solution()
for test in tests:
    input_data = test["input"]
    expected_output = test["output"]
    result = solution.get_decimal_value(input_data)
    
    print(f"Expected Output: {expected_output}, Result: {result}")
    
    assert result == expected_output
```
