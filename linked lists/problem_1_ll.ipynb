# Problem 1
#### https://leetcode.com/problems/convert-binary-number-in-a-linked-list-to-integer/description/

Given <code>head</code> which is a reference node to a singly-linked list. The value of each node in the linked list is either 0 or 1. The linked list holds the binary representation of a number.

Return the decimal value of the number in the linked list.

The most significant bit is at the head of the linked list.

### 1. State the problem clearly. Identify the input & output formats.

Given the start of a linked list traverse the list to find the binary value. Then convert it into decimal form.

Input:
<code>head</code> = the start of a linked list

Output:

<code>decimal</code> value of a binary number

### 2. Come up with some example inputs & outputs. Try to cover all edge cases.

```python
class ListNode:
    def __init__(self, val=0,):
        self.val = val
        self.next = next
```

```python
def create_linked_lists(values):
    if not values:
        return None

    head = ListNode(values[0])
    current = head

    for val in values[1:]:
        current.next = ListNode(val)
        current = current.next
    return head
```

```python
tests = []
```

```python
tests.append({
    "input": create_linked_lists([1, 0, 1]),
    "output": 5
})
```

```python
def get_decimal_value(head: ListNode) -> int:
    current = head
    binary = ""

    while current is not None:
        binary += str(current.val)
        current = current.next

    return int(binary, 2)
```

```python
for test in tests:
    print(get_decimal_value(test["input"]))
    assert get_decimal_value(test["input"]) == test["output"]
```
